```python
import cv2
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
iteso = cv2.imread(r"C:\Users\TAJ\Desktop\Data Lab\my_pic.jpeg")
```

```python
iteso.shape
```

```python
plt.figure()
plt.imshow(iteso)
plt.axis('off')
plt.show()
```

```python
iteso_rgb = cv2.cvtColor(iteso, cv2.COLOR_BGR2RGB)

plt.imshow(iteso_rgb)
plt.axis('off')
plt.show()
```

```python
resized = cv2.resize(iteso, (480, 320))

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
ax1.imshow(iteso_rgb)
ax2.imshow(resized)
ax1.set_title('ITESO')
ax2.set_title('1/2 ITESO')
plt.show()
```

```python
cropped_iteso = iteso_rgb[100:300, 200:800]

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
ax1.imshow(iteso_rgb)
ax2.imshow(cropped_iteso)
ax1.set_title('ITESO Normal')
ax2.set_title('ITESO Cropped')

plt.show()
```
