```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy
```

```python
# Load the selfie (saved in the same directory as this notebook)
selfie = cv2.imread(r"C:\Users\TAJ\Desktop\Data Lab\my_pic.jpeg")
selfie.shape
```

```python
# Display raw (BGR) - colours look off, same as in class
plt.figure()
plt.imshow(selfie)
plt.axis('off')
plt.show()
```

```python
# Fix colours: convert BGR to RGB
selfie_rgb = cv2.cvtColor(selfie, cv2.COLOR_BGR2RGB)

plt.imshow(selfie_rgb)
plt.axis('off')
plt.show()
```

```python

left_eye = selfie_rgb[780:835, 400:520]

print('Left eye coordinates ---->  rows 780:835, cols 400:520')
print('Extracted patch shape:', left_eye.shape)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
ax1.imshow(selfie_rgb)
ax1.set_title('Original Selfie')
ax2.imshow(left_eye)
ax2.set_title('Extracted Left Eye')
plt.show()
```

```python

eye_resized = cv2.resize(left_eye, (300, 120))

print('Original eye shape:', left_eye.shape)     
print('Resized eye shape: ', eye_resized.shape)  

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 5))
ax1.imshow(left_eye)
ax1.set_title('Left Eye - Original size')
ax2.imshow(eye_resized)
ax2.set_title('Left Eye - Resized')
plt.show()
```

```python

selfie_copy = deepcopy(selfie_rgb)


selfie_copy[500:620, 450:750] = eye_resized

fig, ax1 = plt.subplots()
ax1.imshow(selfie_copy)
ax1.set_title('Resized Eye Pasted on Forehead')
plt.show()
```

```python

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

ax1.imshow(selfie_rgb)
ax1.set_title('Original Selfie')

ax2.imshow(selfie_copy)
ax2.set_title('Eye Resized & Pasted on Forehead')

plt.show()
```
